## There is a general decline

Biodiversity and ecosystem health are under serious threat with global declines registered across nearly all major indices, including around 25% of species threatened with extinction across the Animal, Plant and Fungi Kingdoms {cite}`IPBES_2019`. Data overwhelming points to long term and irreversible declines in biodiversity with the term *sixth mass extinction* being commonly evoked to describe the ongoing phenomenon - indeed some experts consider this ongoing loss of ecosystem richness to be a problem of a similar magnitude to that of climate change {cite}`Ceballos_2020`. Among well-studied taxa enough data exists to clearly demonstrate that population, range and diversity have declined precipitously over the last 100 years, for example globally the population of terrestrial mammals has declined by some 80% over the past 75 years {cite}`Wagner_2021`. For understudied taxa like insects that nonetheless make up crucial components of our ecosystems, there is little long term data on historical populations, so a 2017 study that identified a 75% decline between 1989 and 2017 in flying insect mass in nature reserves in Southern Germany caused a media stir and helped to launch conversation about the protection of insects {cite}`Hallman_2017`. 

### One example: Insects in Europe

As Hallman and his coauthors noted however, declines of this magnitude were already in the literature and thus their results should perhaps come as no surprise. To take butterfly populations in Europe, since the best data is generally on butterflies and the best data sets that reach back decades come from select Western European countries, we see a significant general decline {cite}`Warren_2021`. Data from the UK and the Netherlands on butterflies registered a drop of 50% since 1976 and and 1990 respectively, which itself comes after a drop in population on the order of 80% in the Netherlands in the period 1880 - 1940 {cite}`Warren_2021`.  Like most regions, extinction has not occured for nearly all butterfly species in Southern Germany, however the decline in abundance and biomass, the overall population level shifts from specialist sepcies to generalist species, and increasing anthropogenic pressures on increasingly isolated populations leads to the risk that a significant number of extinctions could be on the immediate horizon {cite}`Habel_2019`. While the causes of this decline come from a number of anthropenic pressures, including invasive species, climate change, land-use changes, agricultural intensification, various types of pollution and more, the exact culprit is difficult to identify {cite}`Wagner_2021`.

### The key takeaway: urgent action is needed

Exactly what these results mean for other regions or other taxa remains to be seen, and some insects have demonstrably seen their populations increase. It is certainly possible to debate the results or the methodologies used in evaluating biodiversity when datasets are as patchy as they are for insects, yet the general pattern is clear. The following quotation from a special issue of the Proceedings of the Academy of Natural Sciences on the issue of biodiversity loss in insects sums up the general situation in stark terms {cite}`Wagner_2021`:

>Even without much-needed monitoring and demographic data, enough is already known, based on first principles and records for amphibians, birds, flowering plants, mammals, reptiles, insects, and other taxa, to understand that a biodiversity crisis is accelerating as the planet’s human population grows, increasingly exacerbated by unprecedented recent climate changes and other anthropogenic stressors. Time is not on our side, and urgent action is needed on behalf of nature.

## Switzerland is not an exception

The situation of biodiversity and ecosystem health in Switzerland is no exception to the general trend according to the latest report on biodiversity from the Federal OFfice of the Environment; Widespread extinctions have been avoided, yet urgent action needs to be taken to protect habitats and preserve species across taxonomic groups {cite}`Bafu_2017`. For example, Switzerland has not escaped the pan European decline in insect populations and diversity briefly outlined above: Of the 226 butterfly and moth species monitored nationally (out of some 3700 lepidoptera species confirmed in Switzerland), just over 50% are considered to be under threat or potentially under threat {cite}`Widmer_2021`. Most of this loss occurs due to insecticides, ever increasing intensification of agriculture, and continued artificialisation of land for residential, agricultural or commercial purposes (Ibid).

### Vascular plants

On a national level, like with many other taxa, the vascular plants that are most at risk are aquatic or semi-aquatic {cite}`Bornand_2016`. When grouped by habitat, *milieu naturel*, around 70% of vascular plants associated with freshwater systems were evaluated to be at some level of risk in the 2016 International Union for the Conservation of Nature (IUCN) Red List of species (Ibid). Widespread modification of waterways, the spread of invasive species, run-off from ever intensifying agriculture, artificialisation of shorebanks and regulation of water levels has led to the decline of suitable habitats. Species associated with low altitude prairies, ruderal areas, and cropland/vinyards come next and range from 40% - 60% threatened (Ibid). The threats these species face are primarily a direct result of the changing pratices of agriculture. The species of least concern are associated with alpine habitat, rockfalls, and other largely undisturbed habitat that has been relatively spared from encroachment - although for alpine habitats this is increasingly not the case as agriculture and devleopment continue (Ibid). All of this is unfolding against the backdrop of accelerating climate change that will researchers expect will cause a major shift in habitat distributions that will be a major stress factor for a signficant number of Switzerland's vascular plants {cite}`Vittoz_2013`.

In 2010, the IUCN compiled a comprehensive review of the Red Lists of Switzerland in coordination with the Federal OFfice of the Environment. Unfortunately, at the time, the following conclusion was made:

>Overall,  every  single  Red  List  documents  a  manifest  and  sustained  loss  of  species diversity  at  national  level.  Most  threatened  species  appear  to  be  bound  for  further losses  –  with  no  likelihood  of  this  trend  being  reversed  in  the  short  term.  In  other words,  the  requirement  enshrined  in  the  Swiss  Constitution  that  animal  and  plant   species  be  prevented  from  going  extinct  has  not  been  fulfilled.

10 years later, during the compilation of the latest regional redlist of vascular plants, published in 2019, more than a quarter of Swiss vascular plants are under some threat with *thousands of local extinctions* of sensitive and endangered species identified{cite}`Hicks_2020`:

> Our study presents clear evidence that current efforts to conserve threatened plant species are insufficient to achieve national and international targets (Convention on Biological Diversity (CBD), 2011; Swiss Biodiversity Strategy 2012) for maintaining biodiversity. The current paradigm of protecting and restoring threatened habitats is failing to avert extinctions. Going forward, we need to develop a comprehensive landscape approach, involving the creation of ecological infrastructure and translocation and assisted migration of threatened species into suitable habitats.


## Central Plateau & Jura

The heart of Switzerland's population and economic output, the Central Plateau is also where these localized extinctions are most prominent. Some 60% of the vascular plant species evaluated for the IUCN's Red List are classified as potentially or actually regionally threatened and the Jura region follows with 50% {cite}`Infoflora_2019`. Increased levels of output and consumption, as well as relatively intensive agricultural practices, are identified as the main drivers of risk for regional biodiversity. Not all of this is due directly to higher anthropogenic pressures: the central plateau in particular hosts a large number of habitats for its size that includes isolated, localized populations that are inherently more vulnerable to pressures on their populations.

In the Central Plateau, nearly every group needs attention simply due to the large number of species evaluated as under potential threat or in a more serious risk category. Again, species associated with freshwater habitats are the group of most concern, with some 3/4 of species under potential threat. Around 60% of the remnant alpine populations that survive throughout the plateau, not counting unsustainable populations that occur on the borders of the geographic region, are evaluated as under potential threat, as are just over half of lowland grassland species, and nearly half of forest species {cite}`Infoflora_2019`.

## Preventing the "silent extinction" of genetic diversity

Some actions taken to reinforce biodiversity can in fact cause disruptions to natural ecosystems, especially when the effects are not especially visible. One concern among researchers is the widespread use of commercially grown "wildflowers" that have little genetic diversity {cite}`Johnson_2017`. The primary issue, *genetic erosion*, or loss of genetic diversity within species, is a particular problem among isolated, remnant populations and occurs spontaneously through genetic drift once populations become isolated and poses a particular problem for renaturalization and reintroduction efforts {cite}`Rogers_2004`. The influx of these individuals into the environment on a mass scale can negatively impact the already fragile genetic diversity of fragmented and shrinking wild populations and distort native ranges, although an argument can be made that historical distributions are less critical when facing immminent climate change {cite}`Walker_2020`. As one study of 3 common wildflowers in Switzerland put it, "The results suggest that only plants of relatively local origin should be used in wildflower mixtures, although it is not possible to specify precisely over what distance seed can safely be transferred. The same recommendation is also valid for schemes to reinvigorate endangered plant populations {cite}`Keller_2000`."

This comes back to the definition of biodiversity, and reinforces the fact that regional planners must address diversity on the level of biotopes, species, and genetic diversity within species. While it underscores the need for further research in this field, basic principles direct public and private actors to incorporate the use of regionally native plants (or source-identified) as much as possible into the management of greenspaces, private developments, urban gardens.

## Smart conservation

A judicious use of public spaces, such as road verges, railroad berms and parks, that balances their potential as habitats with the needs and requirements of the city's inhabitants, is one step to mitigate some of the anthropologic pressure. Allowing spaces to go un-mowed, sowing native wild-flower beds, and creating micro-habitats strategically throughout urban and suburban areas can provide measurable benefits to both herbaceous and predatory insects {cite}`Sattler_2011`. In fact, (sub)urban green spaces and road verges can provide plants, insects and birds with suitable conditions to pursue their life cycles and even become a "last refuge" for certain species (ibid). Further, the most recent data of nationwide biodiversity provided by the Biodiversity Monitoring System shows a significant decline in urban vascular plant species {cite}`FOEN_2020`. If this is the case, improvement and maintenance of areas in and around urban centers  could yield outsized results in terms of biodiversity presrvation at this critical juncture for the future of ecosystem services. 

At the same time, research is clear that certain actions are more conducive to sustaining local biodiversity and that their are limits to any one of these relatively convenient strategies. For example, in Finland, reasearchers showed that while road verges can provide suitable refuge habitats for ruderal and some forest species, they do not have much of a capacity to preserve species associated with meadows {cite}`Olden_2021`. Some strategies are clearly strongly dominated in terms of outcomes and inputs needed: plantations of annual wildflower meadows provide only a fraction of the habitat and food resources of perennial meadows, which themselves do not provide resources as continuously throughout the year as the remnant weed populations(CITE). This suggests that molding existing degraded habitats into habitats of moderate quality that resemble natural assemblage is a viable solution to work with nature's changing patterns while giving vulnerable native species a chance to adapt and establish more populations. Indeed, the authors of the 2016 redlist of habitats notes that preserving and upgrading habitats of all types, not just those deemed endangered, can have significant positive effects, just as their loss can have serious negative local implications {cite}`Delarze_2016`.
